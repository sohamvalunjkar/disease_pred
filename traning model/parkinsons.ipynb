```python
pk_dataset['status'] = pd.to_numeric(pk_dataset['status'], errors='coerce')
```

```python
x =pk_dataset.drop(columns=['status'],axis = 1)
y =pk_dataset['status']
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape 
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```
